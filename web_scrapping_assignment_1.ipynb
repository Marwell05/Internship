```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def page(url):
    page1=requests.get(url)
    soup=BeautifulSoup(page1.content)
    return soup
```

# Wikipedia Headers

```python
soup=page('https://en.wikipedia.org/wiki/Main_Page')
```

```python
def head(htag):
    tag=[]
    for val in soup.find_all(htag):
        tag.append(val.text.strip("\n"))
    return tag
```

```python
h1=head("h1")
h2=head("h2")
h3=head("h3")
h4=head("h4")
h5=head("h5")
h6=head("h6")
print("h1 tags:",h1,"\nh2 tags:",h2,"\nh3 tags:",h3,"\nh4 tags:",h4,"\nh5 tags:",h5,"\nh6 tags:",h6)
```

# Top 100 rated imdb movies

```python
soup1=page('https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating')
```

```python
year_top=[]
for date in soup1.find_all('span',class_='lister-item-year text-muted unbold'):
    year_top.append(date.text)
```

```python
ratings_all=[]
for rating in soup1.find_all('div',class_="inline-block ratings-imdb-rating"):
    rating=rating.text.strip("\n")
    ratings_all.append(rating)
```

```python
movie_title=[]
for title in soup1.find_all('h3',class_='lister-item-header'):
    title=title.text.split(".",1)[1]
    title=title.replace("\n","")
    movie_title.append(title)
```

```python
titles=[]
for new in movie_title:
    new=new.split("(")[0]
    titles.append(new)
```

```python
df1=pd.DataFrame({"Name":titles,"Rating":ratings_all,"Year of release":year_top})
df1.index=df1.index+1
df1.index.name="Rank"
df1=df1.style.set_caption("Top 100 rated movies")
df1
```

# Restaurants

```python
soup8=page('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```

```python
restros=[]
for i in soup8.find_all('a',class_='restnt-name ellipsis'):
    restros.append(i.text)
```

```python
loc=[]
for i in soup8.find_all('div',class_='restnt-loc ellipsis'):
    loc.append(i.text)
```

```python
restro_ratings=[]
for i in soup8.find_all('div',class_='restnt-rating rating-4'):
    restro_ratings.append(i.text)
```

```python
image=soup8.find_all('img',class_='no-img')
img_urls=[]
for source in image:
    img_urls.append(source['data-src'])
```

```python
cuisine=[]
for cui in soup8.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(cui.text.split("|")[1])
```

```python
df8=pd.DataFrame({'Restaurant Names':restros,"cuisine":cuisine,'Location':loc,'Ratings':restro_ratings,'Image source':img_urls,})
df8.index=df8.index+1
df8
```
