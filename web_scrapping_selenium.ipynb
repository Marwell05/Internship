# Myntra shoes

```python
driver6=webdriver.Chrome('C:\webdriver\chromedriver.exe') #intialising chrome window
```

```python
driver6.get('https://www.myntra.com/shoes') #opening myntra website with shoes search 
```

```python
select_price=driver6.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
select_price.click() #selecting second filter in price filter column
```

```python
select_color=driver6.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')
select_color.click() #selecting black color from color column
```

```python
#creating list to store scraped data
shoes_brand=[]
shoes_description=[]
shoes_price=[]
```

```python
shoes_brand.clear()
shoes_description.clear()
shoes_price.clear()
```

```python
for i in range(0,2):
    shoe_name=driver6.find_elements_by_xpath("//h3[@class='product-brand']") #scraping brand names
    for shoe in shoe_name:
        shoes_brand.append(shoe.text) #storing brand names in our list for brand names
    shoe_desc=driver6.find_elements_by_xpath("//h4[@class='product-product']") #scraping shoes description
    for desc in shoe_desc:
        shoes_description.append(desc.text) #storing shoes description in our list for description
    shoe_prc=driver6.find_elements_by_xpath("//div[@class='product-price']/span[1]") #scraping shoes price
    for price in shoe_prc:
        price=price.text.strip("Rs.") #removing Rs. from shoes price
        price=price.split("R")[0] #if there is one more Rs. then it will split it based on R and we will store only first index data
        price=price.rjust(len(price)+1,"₹") #enter ruppee symbol infront of prices
        shoes_price.append(price) #storing shoes price in our list for shoes price
    for nxt in range(0,1):
        next_button=driver6.find_element_by_xpath("/html/body/div[2]/div/div[1]/main/div[3]/div[2]/div/div[2]/section/div[2]/ul/li[12]/a")
        next_button.click() #clicking next page
        time.sleep(5)
```

```python
print(len(shoes_price),len(shoes_description),len(shoes_brand)) #checking length of each list
```

```python
df6=pd.DataFrame({"Brand":shoes_brand,"Shoes_description":shoes_description,"Price in Rs.":shoes_price}) #creating dataframe with out scraped data
df6
```

```python
driver6.close() #closing browser
```

```python
driver9=webdriver.Chrome('C:\webdriver\chromedriver.exe') #initialsing chrome browser
```

```python
driver9.get('https://www.ambitionbox.com/') #opening ambitionbox website in our browser
```

```python
search_salary=driver9.find_element_by_xpath('/html/body/div[1]/nav/nav/a[4]') #selecting salary column 
search_salary.click()
```

```python
enter_ds=driver9.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input') #selecting seachbox
enter_ds.send_keys('Data Scientist') #entering Data Scientist in searchbox
time.sleep(2)
```

```python
slt=driver9.find_element_by_xpath("/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p")
slt.click() #clicking on Data Scientist from the suggestions
```

```python
#creating list to store scraped data
company_names_ds=[]
number_of_salaries=[]
avg_salary=[]
min_salary=[]
max_salary=[]
required_experience=[]
```

```python
names=driver9.find_elements_by_xpath("//div[@class='name']/a") #scraping all the company names
for i in names:
    company_names_ds.append(i.text) #adding all the scraped names into our list for company names
```

```python
based=driver9.find_elements_by_xpath("//div[@class='name']/span") #scraping number of people the salaries are based on 
for i in based:
    number_of_salaries.append(i.text)  #adding all the based data in our list
```

```python
exp_req=driver9.find_elements_by_xpath("//div[@class='salaries sbold-list-header']") #scraping required experience
for i in exp_req:
    i=i.text.split(".")[1] #removing unnecessary data
    i=i.replace("\n","") #cleaning scraped data
    required_experience.append(i) #adding required experience to our list
```

```python
avgsal=driver9.find_elements_by_xpath("//p[@class='averageCtc']") #scraping average salaries 
for i in avgsal:
    avg_salary.append(i.text) #adding average salaries to our list
```

```python
min_sal=driver9.find_elements_by_xpath("//div[@class='salary-values']/div[1]")  #scraping minimum salaries 
for i in min_sal: 
    min_salary.append(i.text) #adding minimum salaries to our list
```

```python
max_sal=driver9.find_elements_by_xpath("//div[@class='salary-values']/div[2]") #scraping maximum salaries
for i in max_sal:
    max_salary.append(i.text) #adding maximum salaries to our list
```

```python
print(len(company_names_ds),len(number_of_salaries),len(avg_salary),len(min_salary),len(max_salary),len(required_experience)) #checking length of each column
```

```python
df9=pd.DataFrame({"Company":company_names_ds,"Total salary record":number_of_salaries,"Average_Salary":avg_salary,"Minimum_Salary":min_salary,"Maximum_Salary":max_salary,"Experience Required":required_experience})
df9 #creating dataframe with the scraped data
```
