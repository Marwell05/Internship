# Most viewed videos on YouTube

```python
driver=webdriver.Chrome('C:\webdriver\chromedriver.exe') #initialising chrome browser
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')#opening wikipedia in our browser
```

```python
#creating necessary lists to store our data
rank=[]
name=[]
artist=[]
views=[]
date=[]
all_data=[]
```

# Trending repositories of Github

```python
driver4=webdriver.Chrome('C:\webdriver\chromedriver.exe') #initialising chrome browser
```

```python
driver4.get('https://github.com/') #opening github website in our browser
```

```python
explore=driver4.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details") #scraping explore link
trending=driver4.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]") #scraping trending link
```

```python
actions=ActionChains(driver4) #instantiating ActionChains
```

```python
actions.move_to_element(explore).click().perform() #hovering to explore and clicking
```

```python
actions.move_to_element(trending).click().perform() #moving cursor to trending and clicking it
```

```python
#creating necessary lists
repo_title=[]
repo_desc=[]
language=[]
contributors=[]
repo_link=[]
```

```python
#scraping repository name and its link
repos=driver4.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for val in repos:
    repo_title.append(val.text.split("/")[1]) #splitting data and storing only repository name 
    repo_link.append(val.get_attribute('href')) #storing link of each repository
```

```python
#going to each repo and scraping descreption, contribution and languages used 
for val in repo_link:
    driver4.get(val)
    time.sleep(4)
    try:
        des=driver4.find_element_by_xpath("//p[@class='f4 my-3']")
        repo_desc.append(des.text)
    except:
        repo_desc.append("-")
    try:  #to scrape contributions 4 possible xpaths are there so trying from the most possible xpath to least possible xpath
        contri=driver4.find_element_by_xpath("//div[@class='BorderGrid-row'][5]/div/h2/a/span") #most common xpath
        contributors.append(contri.text)
    except NoSuchElementException:
        try:
            contri=driver4.find_element_by_xpath("//div[@class='BorderGrid-row'][4]/div/h2/a/span")
            contributors.append(contri.text)
        except NoSuchElementException:
            try:
                contri=driver4.find_element_by_xpath("//div[@class='BorderGrid-row'][3]/div/h2/a/span")
                contributors.append(contri.text)
            except NoSuchElementException:
                try:
                    contri=driver4.find_element_by_xpath("//div[@class='BorderGrid-row'][2]/div/h2/a/span") #least common xpath
                    contributors.append(contri.text)
                except: #if none of the path is present then it means there is no contribution column, so appending (-) 
                    contributors.append("-")
    try:  #to scrape languages used 4 possible xpaths are there so trying from the most possible xpath to least possible xpath
        lang=driver4.find_element_by_xpath("//div[@class='BorderGrid-row'][6]/div/ul") #most common xpath
        language.append(lang.text.replace("\n"," "))
    except NoSuchElementException:
        try:
            lang=driver4.find_element_by_xpath("//div[@class='BorderGrid-row'][5]/div/ul")
            language.append(lang.text.replace("\n"," "))
        except NoSuchElementException:
            try:
                lang=driver4.find_element_by_xpath("//div[@class='BorderGrid-row'][4]/div/ul")
                language.append(lang.text.replace("\n"," "))
            except NoSuchElementException:
                try:
                    lang=driver4.find_element_by_xpath("//div[@class='BorderGrid-row'][3]/div/ul") #least common xpath
                    language.append(lang.text.replace("\n"," "))
                except: #if none of the path is present then it means there is no language used column, so appending (-) 
                    language.append("-")
```

```python
print(len(repo_title),len(repo_desc),len(contributors),len(language)) #checking length of each list
```

```python
#creating dataframe with our scraped data
df4=pd.DataFrame({"Repository_title":repo_title,"Description":repo_desc,"Contributors_count":contributors,"Languages_used":language})
df4 
```

```python
driver4.close() #closing our browser
```
